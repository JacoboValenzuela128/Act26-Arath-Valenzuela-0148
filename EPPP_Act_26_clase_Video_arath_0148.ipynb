#### En este programa, la clase Video se utiliza para abstraer un archivo digital, convirtiendo sus características físicas (como el tiempo o la calidad) en datos que la computadora puede procesar.

* Atributos Técnicos: El programa captura el título, la duración (tiempo) y la resolución (calidad de imagen), que son los pilares de cualquier archivo multimedia.

* Interfaz de Usuario: Mediante los métodos leerdatos() y mostrardatos(), el objeto interactúa directamente con el usuario, permitiendo que el código sea dinámico y no estático.

* Instanciación Múltiple: Al crear video1 y video2, demostramos cómo una sola estructura de código puede gestionar una biblioteca de contenidos, manteniendo la información de cada archivo separada y organizada.
![texto del vínculo](https://raw.githubusercontent.com/JacoboValenzuela128/Act26-Arath-Valenzuela-0148/refs/heads/main/Infografiavideo.png)

```python
class Video:
    # El constructor inicializa los atributos
    def __init__(self, titulo="", duracion="", resolucion=""):
        self.titulo = titulo
        self.duracion = duracion
        self.resolucion = resolucion

    # Función para capturar los datos del video
    def leerdatos(self):
        print(f"\n🎥 Configurando nuevo archivo de video...")
        self.titulo = input("Nombre del video: ")
        self.duracion = input("Duración (ej. 10:45 min): ")
        self.resolucion = input("Resolución (ej. 1080p, 4K): ")

    # Función para mostrar la ficha técnica del video
    def mostrardatos(self):
        print(f"\n--- Ficha Técnica: {self.titulo} ---")
        print(f"⏱️ Duración: {self.duracion}")
        print(f"📺 Calidad: {self.resolucion}")
        print("-" * 35)

# --- Creación de los 2 objetos (Instancias) ---

video1 = Video()
video2 = Video()

# --- Ejecución de las funciones ---

# Proceso para el primer video
video1.leerdatos()
video1.mostrardatos()

# Proceso para el segundo video
video2.leerdatos()
video2.mostrardatos()

print("\n-------------------------------------------------------------------")
print(" Programa interactivo creado por Arath Valenzuela (NL52) & Gemini AI")
```
